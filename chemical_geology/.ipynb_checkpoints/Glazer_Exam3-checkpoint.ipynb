## Objective 1
Write a Python routine that takes melt compositions, temperature, $H_{2}O$ and $CO_{2}$ content and calculate the equilibrium pressure.

In Iacono-Marziano et al., we are presented with a model relating melt compositions, $H_{2}O$ and $CO_{2}$ content, temperature, and equilibrium pressure. First, we can look at the equations describing $CO_{2}$ solubility in melt.

### $CO_{2}$ solubility

From Iacono-Marziano et al., we can describe the solubility of carbon dioxide in a melt as follows:

$CO_{2}^{fluid} + O^{2-,melt} \rightarrow CO_{3}^{2-,melt}$; thermodynamic constant $K_C = \frac{a_{CO32-}}{f_{CO2}a_{O2-}} = \frac{\gamma_{CO32-}X_{CO32-}}{f_{CO2}a_{O2-}}$

We can also express the thermodynamic constant in terms of pressure (P), temperature (T), Gibbs free energy ($G^{\circ}$), enthalpy ($H^{\circ}$), entropy ($S^{\circ}$), and volume ($V^{\circ}$).

$ln(K_{C}) = -(\frac{\Delta G^{\circ}}{RT}) = -(\frac{\Delta H^{\circ}}{RT} - \frac{\Delta S^{\circ}}{R} - \frac{P\Delta V^{\circ}}{RT})$

If we combine the two preceeding equations, we get:

$ln(X_{CO32-}) = -ln(\gamma_{CO32-}) + ln(f_{CO2}) + ln(a_{O2-}) - (\frac{\Delta H^{\circ}}{RT} - \frac{\Delta S^{\circ}}{R} - \frac{P\Delta V^{\circ}}{RT})$

If we assume that the carbonate activity coefficient is a linear function of chemical-structural parameters, we can rewrite it as follows:

$-ln(\gamma_{CO32-}) = \Sigma_{i} x_{i} d_{i}$

The $d_{i}$ terms here express chemical control of selected oxides on carbon dioxide solubility, and were determined empirically by the authors and listed in Table 5 of the paper. They are defined in the code cell below. We will be storing these empirical parameters in a dictionary called parameter_dict, indexed by the parameter name and whether we are considering a hydrous or anhydrous basis.

```python
parameter_dict = defaultdict(dict)

parameter_dict['d_H2O']['hydrous'] = -16.4
parameter_dict['d_H2O']['anhydrous'] = 2.3
parameter_dict['d_Al2O3_CaO_K2O_Na2O']['hydrous'] = 4.4
parameter_dict['d_Al2O3_CaO_K2O_Na2O']['anhydrous'] = 3.8
parameter_dict['d_FeO_MgO']['hydrous'] = -17.1
parameter_dict['d_FeO_MgO']['anhydrous'] = -16.3
parameter_dict['d_Na2O_K2O']['hydrous'] = 22.8
parameter_dict['d_Na2O_K2O']['anhydrous'] = 20.1
```

We can also assume the oxygen activity can be expressed as follows:

$ln(a_{O}^{2-}) = b(\frac{NBO}{O})$

Here, NBO/O represents the ratio of non-bridging oxygen to oxygen. The $b$ term defines how carbon dioxide solubility depends on this ratio, and was also determined empirically by the authors of the paper and listed in Table 5. The hydrous and anhydrous values of $b$ are added to our dictionary in the code cell below.

```python
parameter_dict['b_CO2']['hydrous'] = 17.3
parameter_dict['b_CO2']['anhydrous'] = 15.8
```

In the cell below, we define a function to calculate the ratio NBO/O based on the melt composition and whether we are considering a hydrous or anhydrous basis.

```python
def calc_NBO_over_O(molar_fractions, basis):
    # inputs:
        # molar_fractions: Pandas DataArray of molar fractions, with compound names as indices
        # basis: string, either 'hydrous' or 'anhydrous'
    # output:
        # NBO_over_O: float expressing the ratio of non-bridging oxygens to total oxygen
    
    if basis=='hydrous':
        NBO = 2 * (molar_fractions['H2O'] + molar_fractions['K2O'] + molar_fractions['Na2O'] + molar_fractions['CaO'] + molar_fractions['MgO'] + molar_fractions['FeO'] - molar_fractions['Al2O3'])
        NBO_over_O = NBO / (2 * molar_fractions['SiO2'] + 2 * molar_fractions['TiO2'] + 3 * molar_fractions['Al2O3'] + molar_fractions['MgO'] + molar_fractions['FeO']
                            + molar_fractions['CaO'] + molar_fractions['Na2O'] + molar_fractions['K2O'] + molar_fractions['H2O'])
    elif basis=='anhydrous':
        NBO = 2 * (molar_fractions['K2O'] + molar_fractions['Na2O'] + molar_fractions['CaO'] + molar_fractions['MgO'] + molar_fractions['FeO'] - molar_fractions['Al2O3'])
        NBO_over_O = NBO / (2 * molar_fractions['SiO2'] + 2 * molar_fractions['TiO2'] + 3 * molar_fractions['Al2O3'] + molar_fractions['MgO'] + molar_fractions['FeO']
                            + molar_fractions['CaO'] + molar_fractions['Na2O'] + molar_fractions['K2O'])     
    else:
        print("Please enter either 'hydrous' or 'anhydrous' for the basis.")
        
    return NBO_over_O
```

If we assume that enthalpy, entropy, and volume terms don't vary significantly with pressure and temperature, we can combine the equations above to get the following:

$ln([CO_{3}^{2-}]^{ppm}) = \Sigma_{i}x_{i}d_{i} + a \times ln([P_{CO2}]) + b[\frac{NBO}{O}] + \frac{A}{T} + B + C$

A, B, and C are more parameters that have been empirically determined by the authors and listed in Table 5, representing, respectively, the the enthalpy, entropy, and volume changes expressed in J/ppm carbonate. The hydrous and anhydrous values of B and C are added to our dictionary in the code cell below.

```python
parameter_dict['B_CO2']['hydrous'] = -6
parameter_dict['B_CO2']['anhydrous'] = -5.3

parameter_dict['C_CO2']['hydrous'] = 0.12
parameter_dict['C_CO2']['anhydrous'] = 0.14
```

The best fits for $CO_{2}$ from this equation were obtained by the authors using all the parameters defined above, plus the activities of hydrous and anyhydrous $CO_{2}$, defined below.

```python
parameter_dict['a_CO2']['hydrous'] = 1
parameter_dict['a_CO2']['anhydrous'] = 1
```

The final model equation for $CO_{2}$ solubility is therefore:

$ln([CO_{3}^{2-}]^{ppm}) = (x_{H2O} \times d_{H2O} + x_{AI} \times d_{AI} + x_{FeO+MgO} \times d_{FeO+MgO} + x_{Na2O+K2O} \times d_{Na2O+K2O}) + a_{CO2} \times \ln[P_{CO2}] + b_{CO2} \times [\frac{NBO}{O}] + B_{CO2} + C_{CO2} \times \frac{P}{T}$

In this equation, $d_{AI}$ is the same as $d_{Al2O3\_CaO\_K2O\_Na2O}$, and $x_{AI} = \frac{x_{Al2O3}}{x_{CaO} + x_{K2O} + x_{Na2O}}$. The empirical parameters describing $CO_{2}$ solubility are summarized in the table below.

```python
CO2_table = pd.DataFrame(parameter_dict).transpose()
CO2_table
```

Now, we can derive a similar equation for the solubility of $H_{2}O$ in a melt.

### $H_{2}O$ Solubility

Again from Iacono-Marziano et al., we can describe the solubility of $H_{2}O$ in a melt as follows:

$H_{2}O^{fluid} \rightarrow H_{2}O^{melt}$

$H_{2}O^{melt} + O^{2-,melt} \rightarrow 2OH^{-(melt)}$

thermodynamic constant: $K_{H}^{P,T} = \frac{a_{OH-}^{2}}{f_{H2O}\times a_{O2-}} = \frac{(\gamma_{OH-}\times X_{OH-})^{2}}{f_{H2O}\times a_{O2-}}$

We can use the same logic and assumptions as in the $CO_{2}$ solubility section to get the following equation for $H_{2}O$ solubility:

$ln([H_{2}O]^{wt\%}) = \Sigma_{i}x_{i}d_{i} + a \times ln([P_{H2O}]) + b[\frac{NBO}{O}] + \frac{A}{T} + B + C$

Once we plug in the proper empirical parameters, we can express the final model equation for $H_{2}O$ solubility as:

$ln([H_{2}O]^{wt\%}) = a_{H2O} \times ln([P_{H2O}]) + b_{H2O} \times [\frac{NBO}{O}] + B_{H2O} + C_{H2O} \times \frac{P}{T}$

We can now create a new parameter dictionary for the empirically-determined parameters for in this equation, listed in table 6 of the paper.

```python
parameter_dict = defaultdict(dict)

parameter_dict['a_H2O']['hydrous'] = 0.53
parameter_dict['a_H2O']['anhydrous'] = 0.54
parameter_dict['b_H2O']['hydrous'] = 2.35
parameter_dict['b_H2O']['anhydrous'] = 1.24
parameter_dict['B_H2O']['hydrous'] = -3.37
parameter_dict['B_H2O']['anhydrous'] = -2.95
parameter_dict['C_H2O']['hydrous'] = -0.02
parameter_dict['C_H2O']['anhydrous'] = 0.02
```

The empirical parameters describing $H_{2}O$ solubility are summarized in the table below.

```python
H2O_table = pd.DataFrame(parameter_dict).transpose()
H2O_table
```

### Dalton's Law

At this point, we have 2 equations relating $CO_{2}$ and $H_{2}O$ solubility to equilibrium pressure. However, we still have three unknowns: partial pressure of $CO_{2}$ ($P_{CO2}$), partial pressure of $H_{2}O$ ($P_{H2O}$), and overall equilibrium pressure ($P_{eqm}$). In order to solve for all three unknowns, we need a third equation: Dalton's law of partial pressures. This states that in a mixture of non-reacting gases, the equilibrium pressure is the sum of partial pressures of individual gases ($P_{eqm} = P_{CO2} + P_{H2O}$).

### Routine to calculate equilibrium pressure

We can restate our three equations below:

+ $ln([CO_{3}^{2-}]^{ppm}) = (x_{H2O} \times d_{H2O} + x_{AI} \times d_{AI} + x_{FeO+MgO} \times d_{FeO+MgO} + x_{Na2O+K2O} \times d_{Na2O+K2O}) + a_{CO2} \times \ln[P_{CO2}] + b_{CO2} \times [\frac{NBO}{O}] + B_{CO2} + C_{CO2} \times \frac{P_{eqm}}{T}$

+ $ln([H_{2}O]^{wt\%}) = a_{H2O} \times ln([P_{H2O}]) + b_{H2O} \times [\frac{NBO}{O}] + B_{H2O} + C_{H2O} \times \frac{P_{eqm}}{T}$

+ $P_{eqm} = P_{CO2} + P_{H2O}$

We can now use these three equations in a routine to calculate the equilibrium pressure of a solution given a melt composition, temperature, and $H_{2}O$ and $CO_{2}$ contents. By making use of the fsolve functionality from scipy.optimize, we can rewrite these equations to be equal to 0, and solve for the pressure values that make all three equations hold true. The function to do this is defined and annotated below.

```python
def calc_eqm_pressure(vars, melt_comp, temp, wt_percent_H2O, ppm_CO2, basis='anhydrous'):
    # inputs:
        # melt_comp: Pandas DataArray of weight percents, with compound names as indices
        # temp: float, temperature in degrees C
        # wt_percent_H2O: float, concentration of H2O in the solution
        # ppm_CO2: float, concentration of CO2 in the solution
        # basis: string, either 'hydrous' or 'anhydrous'
    # returns:
        # values of P_eqm, P_CO2, and P_H2O that make all three equations above hold true
    
    # defining our three unknowns
    logP_CO2,logP_H2O,P_eqm = vars
    
    # converting from Celsius to Kelvin
    temp = temp + 273
    
    # extracting the empirical parameters for the given basis from each table
    CO2_params = CO2_table[basis]
    H2O_params = H2O_table[basis]
    
    # deriving the molar fractions from the weight percent
    molar_fractions = calc_molar_fractions(melt_comp,wt_percent_H2O,ppm_CO2,basis)
    
    #  calculating the ratio of non-bridging oxygens to total oxygen, using the previously defined function
    NBO_over_O = calc_NBO_over_O(molar_fractions, basis)
    
    # solving for x_AI
    x_AI = molar_fractions.Al2O3 / (molar_fractions.CaO + molar_fractions.K2O + molar_fractions.Na2O)
    
    # CO2 solubility equation, set equal to 0
    if(ppm_CO2==0):
        eq1 = ((molar_fractions.H2O * CO2_params.d_H2O + x_AI * CO2_params.d_Al2O3_CaO_K2O_Na2O + (molar_fractions.FeO + molar_fractions.MgO) * CO2_params.d_FeO_MgO +
           (molar_fractions.Na2O + molar_fractions.K2O) * CO2_params.d_Na2O_K2O) + CO2_params.a_CO2 * logP_CO2 +
           CO2_params.b_CO2 * NBO_over_O + CO2_params.B_CO2 + CO2_params.C_CO2 * P_eqm / temp)
    else:
        eq1 = ((molar_fractions.H2O * CO2_params.d_H2O + x_AI * CO2_params.d_Al2O3_CaO_K2O_Na2O + (molar_fractions.FeO + molar_fractions.MgO) * CO2_params.d_FeO_MgO +
           (molar_fractions.Na2O + molar_fractions.K2O) * CO2_params.d_Na2O_K2O) + CO2_params.a_CO2 * logP_CO2 +
           CO2_params.b_CO2 * NBO_over_O + CO2_params.B_CO2 + CO2_params.C_CO2 * P_eqm / temp) - np.log(ppm_CO2)

    # H2O solubility equation, set equal to 0
    eq2 = (H2O_params.a_H2O * logP_H2O + H2O_params.b_H2O * NBO_over_O + H2O_params.B_H2O +
           H2O_params.C_H2O * (P_eqm / temp)) - np.log(wt_percent_H2O)
    
    # Dalton's law equation, set equal to 0
    eq3 = P_eqm - np.e**(logP_CO2) - np.e**(logP_H2O)
    
    return [eq1, eq2, eq3]
```

The function below converts the melt composition in weight percent to molar fractions, including $H_{2}O$ if the basis is hydrous.

```python
def calc_molar_fractions(melt_comp, wt_percent_H2O, ppm_CO2, basis):
    # inputs:
        # melt_comp: Pandas DataArray of weight percents, with compound names as indices
        # wt_percent_H2O: float, concentration of H2O in the solution
        # ppm_CO2: float, concentration of CO2 in the solution
        # basis: string, either 'hydrous' or 'anhydrous'
    # returns:
        # molar_fractions: Pandas DataArray of molar fractions, with compound names as indices
    
    denominator = (melt_comp.SiO2/60.08 + melt_comp.TiO2/79.87 + melt_comp.Al2O3/101.96 + melt_comp.FeO/71.84 +
                   melt_comp.MgO/40.30 + melt_comp.CaO/56.08 + melt_comp.Na2O/61.98 + melt_comp.K2O/94.20 + 
                   melt_comp.P2O5/283.89) + (ppm_CO2/10000)/44.01
    
    if(basis=='hydrous'):
        denominator = denominator + wt_percent_H2O/18.02
        
    molar_fractions = defaultdict(dict)
    molar_fractions['SiO2'] = (melt_comp.SiO2 / 60.08) / denominator
    molar_fractions['TiO2'] = (melt_comp.TiO2 / 79.87) / denominator
    molar_fractions['Al2O3'] = (melt_comp.Al2O3 / 101.96) / denominator
    molar_fractions['FeO'] = (melt_comp.FeO / 71.84) / denominator
    molar_fractions['MgO'] = (melt_comp.MgO / 40.30) / denominator
    molar_fractions['CaO'] = (melt_comp.CaO / 56.08) / denominator
    molar_fractions['Na2O'] = (melt_comp.Na2O / 61.98) / denominator
    molar_fractions['K2O'] = (melt_comp.K2O / 94.20) / denominator
    molar_fractions['P2O5'] = (melt_comp.P2O5 / 283.89) / denominator
    molar_fractions['H2O'] = (melt_comp.H2O_wt_percent / 18.02) / denominator
    
    molar_fractions = pd.DataFrame(molar_fractions.values(),index=molar_fractions.keys())
    
    return molar_fractions[0]
```

### Testing model

We can test out our routine using the data presented in tables 1 and 2 of Iacono-Marziano et al.. Table 1 gives us the composition of a starting glass of an alkali-basalt from Mt. Etna, Italy. Table 2 gives us the experimental conditions and results of the experiments with this basalt. The measured compositions for the starting glass and the experimental conditions are presented in the following data tables.

```python
Etna_data = {'Starting_Glass':[47.95,1.67,17.32,10.24,5.76,10.93,3.45,1.99,.51,99.82,.010]}
melt_comps = pd.DataFrame(Etna_data,index=['SiO2','TiO2','Al2O3','FeO','MgO','CaO','Na2O','K2O',
                                      'P2O5','total','H2O_wt_percent'])

exp_params = {'P_eqm_bars':[485,485,485,485,1015,1015,1015,1017,1017,1530,1530,1530,1530,2047,2047,2055,2055,2135,2135,2754,3080,3080,4185,4185,4185,4185],
                           'wt_percent_H2O':[0.95,1.18,2.22,2.02,0.99,2.12,3.48,0.8,2.08,1.43,3.01,4.46,4.87,1.04,2.82,0.87,3.45,2.70,5.09,1.64,1.09,3.54,1.36,1.47,3.31,2.16],
                           'ppm_CO2':[306,191,0,0,843,548,0,808,534,1278,1035,0,0,1853,1489,1706,1408,1412,0,2515,2816,2416,3673,3965,4061,4230]}
experimental_conditions = pd.DataFrame(exp_params)

display(melt_comps)
display(experimental_conditions)
```

Now, we use a for-loop to cycle through each of the experimental conditions, using the weight percent $H_{2}O$ and ppm $CO_{2}$ as inputs, and seeing if the equilibrium pressure returned by the model matches the equilibrium pressure of the experiment. As stated in the paper, we use a temperature of 1200 degrees Celsius, and calculate in a an anhydrous basis.

```python
melt_comp = melt_comps.Starting_Glass
temp = 1200
basis = 'anhydrous'

results_dict = defaultdict(dict)

for i in np.arange(experimental_conditions.shape[0]):
    wt_percent_H2O = experimental_conditions.loc[i,'wt_percent_H2O']
    ppm_CO2 = experimental_conditions.loc[i,'ppm_CO2']
    logP_CO2, logP_H2O, P_eqm = fsolve(calc_eqm_pressure, [200,100,300], args=(melt_comp,temp,wt_percent_H2O,ppm_CO2,basis))
    results_dict[i]['Iacono-Marziano P_eqm (bars)'] = experimental_conditions.loc[i,'P_eqm_bars']
    results_dict[i]['modeled P_eqm (bars)'] = P_eqm
    results_dict[i]['modeled P_CO2 (bars)'] = np.e**logP_CO2
    results_dict[i]['modeled P_H2O (bars)'] = np.e**logP_H2O
    
results_table = pd.DataFrame(results_dict).transpose()

results_table
```

The scatter plot below shows the modeled equilibrium pressures versus the actual experimental equilibrium pressures.

```python
fig,ax=plt.subplots()

slope, intercept, r_value, p_value, std_err = linregress(results_table['Iacono-Marziano P_eqm (bars)'], results_table['modeled P_eqm (bars)'])

results_table.plot.scatter(x='Iacono-Marziano P_eqm (bars)', y='modeled P_eqm (bars)',ax=ax,color='blue')

best_fit_line_label = 'best-fit line, slope = ' + str(round(slope,3))

plt.plot([0,5000],[0,5000],color='black',linestyle='dashed',label='x=y line')
plt.plot([0,5000],[intercept,5000*slope+intercept],color='blue',linestyle='dashed',label=best_fit_line_label)

plt.xlim(0,5000)
plt.ylim(0,5000)

plt.legend()
plt.title('modeled equilibrium pressure vs. actual (experimental) equilibrium pressure')

plt.show();

print('r = ', str(r_value))
```

As you can see here, our model does a good job of predicting the experimental pressures! The x=y line is shown for comparison. The best-fit line to our data has a slope of 1.03, with a correlation coefficient r of 0.98.

## Objective 2
Write a Python routine that takes melt compositions, temperature, and starting pressure and calculates the $H_{2}O$ and $CO_{2}$ content from initial pressure up to 0.1MPa.

We can modify our solver function from objective 1 to return $H_{2}O$ and $CO_{2}$ concentrations for a given equilibrium pressure. One complication is that we now have four unknowns: $P_{CO2}$, $P_{H2O}$, wt_percent_H2O, and ppm_CO2, and still just the same three equations from earlier, restated below for convenience:

+ $ln([CO_{3}^{2-}]^{ppm}) = (x_{H2O} \times d_{H2O} + x_{AI} \times d_{AI} + x_{FeO+MgO} \times d_{FeO+MgO} + x_{Na2O+K2O} \times d_{Na2O+K2O}) + a_{CO2} \times \ln[P_{CO2}] + b_{CO2} \times [\frac{NBO}{O}] + B_{CO2} + C_{CO2} \times \frac{P_{eqm}}{T}$

+ $ln([H_{2}O]^{wt\%}) = a_{H2O} \times ln([P_{H2O}]) + b_{H2O} \times [\frac{NBO}{O}] + B_{H2O} + C_{H2O} \times \frac{P_{eqm}}{T}$

+ $P_{eqm} = P_{CO2} + P_{H2O}$

To remedy this, we can make use of another parameter provided in the experimental conditions table (Table 2) in Iacono-Marziano et al. We can set the molar fraction of $CO_{2}$ in the gas phase as the ratio of $P_{CO2}$ to $P_{eqm}$. This gives us a fourth equation:

+ $P_{CO2} = X_{CO2} \times P_{eqm}$

Here, we add $X_{CO2}$ to the experimental conditions table.

```python
experimental_conditions['X_CO2'] = [.68,.56,0,0,.84,.49,0,.88,.51,.81,.43,0,0,.9,.61,.93,.48,.65,0,.87,.94,.64,.94,.93,.76,.87]
experimental_conditions
```

Now, we are ready to define our function below. Note that this function will only work with an anhydrous basis, as we don't have the weight percent of $H_{2}O$ as an input.

```python
def calc_H2O_CO2(vars, melt_comp, temp, P_eqm, X_CO2):
    # inputs:
        # melt_comp: Pandas DataArray of weight percents, with compound names as indices
        # temp: float, temperature in degrees C
        # P: float, starting pressure in bar (defaults to 1 bar, or 0.1 MPa)
        # X_CO2: float, molar fraction of CO2 in gas phase
    # returns:
        # values of wt_percent_H2O and ppm_CO2 that make all three equations above hold true

    # defining our four unknowns
    logwt_percent_H2O, logppm_CO2, logP_CO2, logP_H2O = vars
    
    # converting from Celsius to Kelvin
    temp = temp + 273
    
    # extracting the empirical parameters for the given basis from each table
    CO2_params = CO2_table[basis]
    H2O_params = H2O_table[basis]
    
    # deriving the molar fractions from the weight percent
     ### note that we are inputting the concentration of H2O as 0, because we are using an anhydrous basis,
        ### and also approximating the ppm_CO2 as 0
    molar_fractions = calc_molar_fractions(melt_comp,0,0,'anhydrous')
    
    #  calculating the ratio of non-bridging oxygens to total oxygen, using the previously defined function
    NBO_over_O = calc_NBO_over_O(molar_fractions, 'anhydrous')
    
    # solving for x_AI
    x_AI = molar_fractions.Al2O3 / (molar_fractions.CaO + molar_fractions.K2O + molar_fractions.Na2O)
    
    # CO2 solubility equation, set equal to 0
    eq1 = ((molar_fractions.H2O * CO2_params.d_H2O + x_AI * CO2_params.d_Al2O3_CaO_K2O_Na2O + (molar_fractions.FeO + molar_fractions.MgO) * CO2_params.d_FeO_MgO +
        (molar_fractions.Na2O + molar_fractions.K2O) * CO2_params.d_Na2O_K2O) + CO2_params.a_CO2 * logP_CO2 +
        CO2_params.b_CO2 * NBO_over_O + CO2_params.B_CO2 + CO2_params.C_CO2 * P_eqm / temp) - logppm_CO2

    # H2O solubility equation, set equal to 0
    eq2 = (H2O_params.a_H2O * logP_H2O + H2O_params.b_H2O * NBO_over_O + H2O_params.B_H2O +
           H2O_params.C_H2O * (P_eqm / temp)) - logwt_percent_H2O
    
    # Dalton's law equation, set equal to 0
    eq3 = P_eqm - np.e**(logP_CO2) - np.e**(logP_H2O)
    
    # proportion of CO2 in gas phase
    eq4 = X_CO2 * P_eqm - np.e**(logP_CO2)
    
    return [eq1, eq2, eq3, eq4]
```

Now, we use a for-loop to cycle through each of the experimental conditions, using the experimental equilibrium pressure and $X_{CO2}$ as inputs, and seeing if the concentrations of $H_{2}O$ and $CO_{2}$ returned by the model match the experimental conditions. As stated in the paper, we use a temperature of 1200 degrees Celsius, and calculate in a an anhydrous basis.

```python
melt_comp = melt_comps.Starting_Glass
temp = 1200

results_dict_2 = defaultdict(dict)

for i in np.arange(experimental_conditions.shape[0]):
    
    X_CO2 = experimental_conditions.loc[i,'X_CO2']
    P_eqm = experimental_conditions.loc[i,'P_eqm_bars']
    logwt_percent_H2O, logppm_CO2, logP_CO2, logP_H2O = fsolve(calc_H2O_CO2,[0,0,10,10],args=(melt_comp,temp,P_eqm,X_CO2))

    results_dict_2[i]['Iacono-Marziano wt_percent_H2O'] = experimental_conditions.loc[i,'wt_percent_H2O']
    results_dict_2[i]['Iacono-Marziano ppm_CO2'] = experimental_conditions.loc[i,'ppm_CO2']
    results_dict_2[i]['wt_percent_H2O'] = np.e**logwt_percent_H2O
    results_dict_2[i]['ppm_CO2'] = np.e**logppm_CO2
    
results_table_2 = pd.DataFrame(results_dict_2).transpose()
results_table_2
```

The scatter plots below show the modeled $CO_{2}$ and $H_{2}O$ concentrations versus the actual experimental concentrations.

```python
fig,ax=plt.subplots(1,2,figsize=(16,6))

results_table_2.plot.scatter(x='Iacono-Marziano ppm_CO2',y='ppm_CO2',ax=ax[0],color='blue')
results_table_2.plot.scatter(x='Iacono-Marziano wt_percent_H2O',y='wt_percent_H2O',ax=ax[1],color='orange')

CO2slope, CO2intercept, CO2r_value, CO2p_value, CO2std_err = linregress(results_table_2['Iacono-Marziano ppm_CO2'], results_table_2['ppm_CO2'])
CO2_best_fit_line_label = 'best-fit line, slope = ' + str(round(CO2slope,3))
ax[0].plot([0,4500],[0,4500],color='black',linestyle='dashed',label='x=y line')
ax[0].plot([0,4500],[CO2intercept,4500*CO2slope+CO2intercept],color='blue',linestyle='dashed',label=CO2_best_fit_line_label)
ax[0].legend()
ax[0].set_xlim(0,4500)
ax[0].set_ylim(0,4500)
ax[0].set_title('modeled CO2 concentration vs. actual (experimental) CO2 concentration')

H2Oslope, H2Ointercept, H2Or_value, H2Op_value, H2Ostd_err = linregress(results_table_2['Iacono-Marziano wt_percent_H2O'], results_table_2['wt_percent_H2O'])
H2O_best_fit_line_label = 'best-fit line, slope = ' + str(round(H2Oslope,3))
ax[1].plot([0,5],[0,5],color='black',linestyle='dashed',label='x=y line')
ax[1].plot([0,5],[H2Ointercept,5*H2Oslope+H2Ointercept],color='orange',linestyle='dashed',label=H2O_best_fit_line_label)
ax[1].legend()
ax[1].set_xlim(0,5)
ax[1].set_ylim(0,5)
ax[1].set_title('modeled H2O concentration vs. actual (experimental) H2O concentration')

fig.show();

print('ppm_CO2 r = ', str(CO2r_value))
print('ppm_H2O r = ', str(H2Or_value))
```

As you can see here, our model does a good job of predicting the $CO_{2}$ and $H_{2}O concentrations! The x=y lines are shown for comparison. The best-fit lines to our data have slopes of 0.90 and 0.81 with correlation values of 0.98 and 0.99, respectively.
